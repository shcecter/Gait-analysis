```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import approximating as app
import gait_pca as gpca
import normal_gait as NG


%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def pca_res(sheetname):
    joints_l = app.load_data(sheetname=sheetname)
    pca_res = gpca.gait_ftransform(joints_l, n_components=4)
    return pca_res, joints_l
    
```

```python
bask_res = pca_res('baskakova')
gpca.plot_gait_pca(*bask_res)
```

```python
baskf_res = pca_res('baskakova forced')
gpca.plot_gait_pca(*baskf_res)
```
