```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


%matplotlib inline
```

```python
import approximating as app
```

```python
sheetname = 'femoral hernia'
joints_l = app.load_data(sheetname)
```

```python
r_hip = joints_l[0]
r_knee = joints_l[1]
```

```python
input_matrix = np.vstack((joints_l))
pca = PCA(n_components=1)
pca.fit(input_matrix)
result = pca.transform(input_matrix)
```

```python
result.shape
```

```python
output_matrix = pca.inverse_transform(result)
out_r_hip = output_matrix[0, :]
```

```python
output_matrix.shape
```

```python
out_r_hip.shape
```

```python
plt.plot(out_r_hip, 'b', r_hip, 'r')
plt.grid()
```

```python
out_r_knee = output_matrix[1, :]
plt.plot(out_r_knee, 'b', r_knee, 'r')
plt.grid()
```

```python
for i in range(output_matrix.shape[0]):
    plt.plot(output_matrix[i, :])
    
    
plt.grid()
```

```python
pca_n2 = PCA(n_components=2)
trans = pca_n2.fit_transform(input_matrix)
out_n2 = pca_n2.inverse_transform(trans)
```

```python
for i in range(out_n2.shape[0]):
    plt.plot(out_n2[i, :])
    
    
plt.grid()
```
