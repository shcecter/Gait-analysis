```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import approximating as app
import normal_gait as NG


%matplotlib inline
```

```python
sheetname = 'femoral hernia'
joints_l = app.load_data(sheetname=sheetname)
r_hip = joints_l[0]
r_knee = joints_l[1]
```

```python
l_hip = joints_l[3]
l_knee = joints_l[4]
```

```python
n2_hip, n2_knee = NG.load_normal_gait(2)
```

```python
r_phase_angle = np.rad2deg(np.arctan(r_hip / r_knee))
n_phase_angle = np.rad2deg(np.arctan(n2_hip / n2_knee))
```

# Phase angles of pathological and normal walking

```python
plt.plot(r_phase_angle, 'r', n_phase_angle, 'b')
plt.grid()
```

```python
r_magnitude = np.sqrt(np.square(r_hip) + np.square(r_knee))
n_magnitude = np.sqrt(np.square(n2_hip) + np.square(n2_knee))
```

# Phase magnitudes of pathological and normal walking

```python
plt.plot(r_magnitude, 'r', n_magnitude, 'b')
plt.grid()
```

```python
normal_l = [NG.load_normal_gait(i + 1) for i in range(3)]
```

# Several noramal walking trials analysis

```python
def get_phase_angles(x, y):
    rad = np.arctan(x / y)
    grad = np.rad2deg(rad)
    return grad
```

```python
def get_phase_magnitudes(x, y):
    sum_of_squares = np.square(x) + np.square(y)
    magnitudes = np.sqrt(sum_of_squares)
    return magnitudes
```

```python
for joint in normal_l:
    plt.figure(1)
    phase_angles = get_phase_angles(*joint)
    plt.plot(phase_angles)
    plt.figure(2)
    magnitudes = get_phase_magnitudes(*joint)
    plt.plot(magnitudes)
    
plt.figure(1)
plt.grid()
plt.figure(2)
plt.grid()
```

# Phase angle versus magnitude

```python
plt.figure(1)
plt.plot(r_phase_angle, r_magnitude, 'r')
plt.grid()
plt.figure(2)
plt.plot(n_phase_angle, n_magnitude, 'b')
plt.grid()
```

# Several normal walking trials

```python
for joint in normal_l:
    plt.plot(get_phase_angles(*joint), get_phase_magnitudes(*joint))
    
    
plt.grid()
```

# Left and right pathological walking (femoral hernia)

```python
plt.plot(get_phase_angles(l_hip, l_knee), get_phase_magnitudes(l_hip, l_knee))
plt.plot(get_phase_angles(r_hip, r_knee), get_phase_magnitudes(r_hip, r_knee))
plt.grid()
plt.figure()
plt.plot(r_knee, r_hip, 'r', l_knee, l_hip, 'b')
plt.grid()
```
